```python
# concatenate and merge DataFrame(JOINS)
```

```python
df1 = pd.DataFrame({"ID":[1,2,3],'Name':['A','B','C']})
df1
```

```python
df2 = pd.DataFrame({'ID' :[1,2,3,4],'Score':[33,54,64,72]})
df2
```

```python
pd.concat([df1,df2],axis=0) # vertical / row level / top on top
```

```python
pd.concat([df1,df2],axis=1) # horizantal / row level / side on side
```

```python
# merge = join
pd.merge(df1, df2, how="inner", on='ID')
```

```python
pd.merge(df1, df2, how='outer', on='ID')
```

```python
pd.merge(df1, df2, how='left', on='ID')
```

```python
pd.merge(df1, df2, how='outer', left_on='ID',right_on = 'ID')
```
