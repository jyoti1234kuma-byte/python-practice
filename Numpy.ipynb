```python
# 5. Array reshaping and flattening

```

```python
# Reshape Array: changing it's dimension (e.g., row and columns) without altering the actual data
# Flatten array: transformational a multi-dimensional array into a single-dimensional array

```

```python
# shape = (1,6) - 1 row and 6 column
arr1 = np.array([1,2,3,4,5,6])
print(arr1)
```

```python
# shape = (6,1) -6 rows and 1 column
reshaped = arr1.reshape((6,1))
print(reshaped)
```

```python
# reshaped array from (1,6) to (2,3)

reshaped2 = arr1.reshape((2,3))
print(reshaped2)
```

```python
# flattened array from higher (2D) dimensional (2,3) to (1,6) (1D)
print(reshaped2.flatten())


```

```python
#6. Array Stacking and Splitting
```

```python
# Array Stacking: combining multple arrays aling a new axis, resulting in an array with a higher dimension than the input arrays
# Array Splitting:
```

```python
# Staking 
a = np.array([1,2,3])
b = np.array([4,5,6])

# vertical stackling - row wise  -> output is 2D
print(np.vstack((a,b)))

# horizantal stacking - column wise
print(np.hstack((a)))
```

```python
# splitting
c = np.array([[1,2,3],[4,5,6]])
print(c)
```

```python
# horizontal split -> input and output aray are same dimention

hsplit = np.hsplit(c,3)
for s in hsplit:
    print(s)
```

```python
# bertical split -> input and output array are same dimention
vsplit = np.vsplit(c,2)

for s in vsplit:
    print(s)
```

```python
# 11. Broadcasting

# Broadcasting: is a mechanism that allow arithmetic to be performed on arrays of different shapes and sizes,
# it eliminates the need for explicit loops or reshaping operating, making code more concise, readable and efficient.

# Broadcasting Rules:
   # comparing shapes from right to left.
   # Dimensions must match or be 1.
   # If one array has fewer dimensions, pad its shape on the left with is.
   # Any dimension equal to 1 can stretch to match the other array.
   
```

```python
a = np.array([1,2,3,4])
b = np.array([5])
print(a+b)       # adding 1D with One-element Array to 1D with multi-element array
```

```python
c = np.array([[1,2,3],[4,5,6]])
d = np.array([10,20,30])
print(c+d)            # adding 2D with 6-element array(2,3) to 1D with 6-element array
```
